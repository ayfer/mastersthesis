```python
import pandas as pd

splits = {'train': 'imbalanced/train-00000-of-00001.parquet', 'validation': 'imbalanced/validation-00000-of-00001.parquet', 'test': 'imbalanced/test-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/clinc/clinc_oos/" + splits["train"])
```

```python
bank_intents = [109, 35, 68,89,87]
df_bank = df[df["intent"].isin(bank_intents)]
```

```python
df_notbank = df[~df["intent"].isin(bank_intents)]
```

```python
df_notbank = df_notbank.sample(25).reset_index(drop=True)
```

```python
df_bank_109 = df_bank[df_bank["intent"] == 109]
df_bank_35 = df_bank[df_bank["intent"] == 35]
df_bank_68 = df_bank[df_bank["intent"] == 68]
df_bank_89 = df_bank[df_bank["intent"] == 89]
df_bank_87 = df_bank[df_bank["intent"] == 87]
```

```python


dfs={"CreditLimitChange":df_bank_109[["text"]], "AccountBalance":df_bank_35[["text"]], "Transfer":df_bank_68[["text"]], "PinChange":df_bank_89[["text"]], "DamagedCard":df_bank_87[["text"]], "Unknown":df_notbank[["text"]]}
```

```python
with pd.ExcelWriter('/Users/a.aslan/Desktop/IR_LC/in/in_bank.xlsx') as writer:
    for sheet_name, df in dfs.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
